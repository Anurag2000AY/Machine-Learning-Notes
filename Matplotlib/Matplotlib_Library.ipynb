# Lec 4. Anatomy Of A Matplotlib Figure

### Matplotlib example workflow

```python
# 0. import matplotlib and get it ready for plotting in Jupter
%matplotlib inline
import matplotlib.pyplot as plt

# 1. Prepare data
x= [1,2,3,4]
y=[11,22,33,44]

# 2. Setup Plot
fig, ax= plt.subplots(figsize=(10,10)) # width height of the fig

# 3. Plot Data
ax.plot(x,y)

# 4 Customise Plot
ax.set(title = "Simple Plot",
        xlabel= "x-axis",
        ylabel= "y-axis")

# 5 save & show (you save the whole figure)
fig.savefig("sample-plot.png")
```

### Two options for subplots

```python
# subplots option 1
# it helpful in looking different infomation in graphical form within one fig

fig, ((ax1,ax2), (ax3,ax4))=plt.subplots(nrows=2,
                                       ncols=2,
                                       figsize=(10,5))
ax1.plot(x,x/2) #requries x and y
ax2.scatter(np.random.random(10), np.random.random(10)) # require x and y
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values()) # require x and y
ax4.hist(np.random.randn(1000)); #dosen't require x and y
```

```python
# back to problem
# adding date cloumn to our data frame

# note new colums length must be equal to number of rows in your dataframe
car_sales["Sales_Date"] = pd.date_range("21/5/2022", periods=len(car_sales))
car_sales
```

```python
type(car_sales["Price"][0])
```

it happens becuse we have price column in string format hece here cumsum operation means cummulative of string means concatenation so we need to convert into interger using astype fuction see below

```python
# Makin column for representing total sales upto that date
car_sales["Total Sales"]= car_sales["Price"].astype(int).cumsum()
car_sales
```

```python
# Lets plot tatal sales over time
 
car_sales.plot(x="Sales_Date", y="Total Sales");
```
